```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
#importing datasets
df=pd.read_csv("train.csv")
```

```python
df
```

```python
df.head(5000)
```

```python
types = ['toxic', 'severe_toxic', 'obscene', 'threat',
       'insult', 'identity_hate']

sampled_train1 = df[(df['toxic'] ==0) & (df['severe_toxic'] ==0) & (df['obscene'] ==0)
                       & (df['threat'] ==0) & (df['insult'] ==0) & (df['identity_hate'] ==0)][:20000]

sampled_train2 = df[(df['toxic'] !=0) | (df['severe_toxic'] !=0) | (df['obscene'] !=0) 
                | (df['threat'] !=0) | (df['insult'] !=0) | (df['identity_hate'] !=0)][:10000]
```

```python
sampled_train = sampled_train2.append(sampled_train1)
sampled_train = sampled_train.sample(frac=1)
```

```python
sampled_train.head(30000)
```

```python
toxic=sampled_train[:][:30000]
```

```python
toxic.reset_index(inplace = True) 
  
```

```python
toxic
```

```python
toxic['comment_text']
```

```python
toxic_labels = toxic[["toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate"]]
toxic_labels.head(15000)
```

```python
import re
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from string import punctuation
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()
from textblob import TextBlob
X=[]
for i in range(0,30000):
       comment=toxic["comment_text"][i]
       comment = re.sub('[^a-zA-Z]', ' ', comment)
       comment = comment.lower()
       comment= comment.split()
       comment = [ps.stem(word) for word in comment if not word in set(stopwords.words('english'))]
       comment = ' '.join(comment)
       X.append(comment)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize 
token=[]
for i in X:
    token.append(word_tokenize(i))
    
```

# Bag of words

```python
        
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1500,stop_words='english')
X1 = cv.fit_transform(X).toarray()
```

```python
ip_dim = X1.shape[1]
```

```python
ip_dim
```

```python
X1.shape
```

```python
y = toxic_labels.values
```

```python
y.shape
```

# Train and test split the data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X1, y, test_size=0.20, random_state=42)
```

# ANN Model

```python
from keras.models import Sequential
from keras.layers import Dense
```

# Adding input layer,hidden layer and output layer

```python
model = Sequential()

model.add(Dense(output_dim = 64, init = 'uniform', activation = 'relu', input_dim = ip_dim))#input layer

model.add(Dense(output_dim = 32, init = 'uniform', activation = 'relu'))#hidden layer

model.add(Dense(output_dim = 6, init = 'uniform', activation = 'sigmoid'))#output layer
```

# Compliling the model

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

# Train and Test the model 


```python
model.fit(X_train,y_train,batch_size=32,epochs=50)
```

# optimize the model

```python
y_pred=model.predict(X_test)

y_pred=y_pred>0.5
```

```python
y_pred
```

```python
y_test
```

```python
score = model.evaluate(X_test, y_test)

print("Test Score:", score[0])
print("Test Accuracy:", score[1])
```
